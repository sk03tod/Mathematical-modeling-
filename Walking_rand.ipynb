```python
import numpy as np 
import matplotlib.pyplot as plt
import random as rnd
```

```python
Walkingarea = 20
l = 0.2
def Gauss(x, nstep, ntr):
    return ((2*ntr)/(np.sqrt(2*np.pi*nstep)))*np.exp(-((Walkingarea/2)-x)**2/(2*nstep*l**2))

def Gauss_array(ns, nt):
    return list(Gauss(i*l, ns, nt) for i in range(int(Walkingarea/l)))
```

```python
Xgrid = list(round(0.2*i, 2) for i in range(int(Walkingarea/l)))
print(Xgrid)
```

```python
def Random_walking(ntr, nstp):
    array = list(0 for i in range(int(Walkingarea/l)))
    for i in range(ntr):
        counter = int(Walkingarea/2/l)
        for j in range(nstp):
            if rnd.random()>0.5:
                counter+=1
            else: counter-=1
        array[counter]+=1
    return list(x/ntr for x in array)
```

```python
plt.plot(Xgrid, Random_walking(100000, 90))
plt.plot(Xgrid, Gauss_array(90, 1))
plt.show()
plt.plot(Xgrid, Random_walking(100000, 100))
plt.plot(Xgrid, Gauss_array(100, 1))
plt.show()
plt.plot(Xgrid, Random_walking(100000, 20))
plt.plot(Xgrid, Gauss_array(20, 1))
plt.show()
plt.plot(Xgrid, Random_walking(100000, 50))
plt.plot(Xgrid, Gauss_array(50, 1))
plt.show()
```
