```python
#Блок констант
k1 = 0.1
antik1 = 0.1
k2 = 0.15
k3 = 5

#Функції швидкостей
def Svel(s, e0, se):
    sveloc=2*antik1*se-2*k1*(e0-se)*s*2
    return sveloc
def SEvel(s, e0, se):
    seveloc=k1*(e0-se)*s*2-antik1*se-k2*se
    return seveloc


```

# **Реалізуємо методом Рунге Кутта 4 порядку**

```python
Sub = 1
SubE = 0
Eaxis = []
Paxis = []
Saxis=[]
SEaxis=[]
taxis=list(i/10000 for i in range(0, 1000000))
Witer = 0
Riter = 0
CounterW = 0
CounterR = 0
h = 0.0001
for i in range(1000000):
    Riter = h*Svel(Sub, 1, SubE)
    Witer = h*SEvel(Sub, 1, SubE)
    CounterW+=Witer
    CounterR+=Riter
    Riter = h*Svel(Sub+Riter/2, 1, SubE+Witer/2)
    Witer = h*SEvel(Sub+Riter/2, 1, SubE+Witer/2)
    CounterR+=2*Riter
    CounterW+=2*Witer
    Riter = h*Svel(Sub+Riter/2, 1, SubE+Witer/2)
    Witer = h*SEvel(Sub+Riter/2, 1, SubE+Witer/2)
    CounterR+=2*Riter
    CounterW+=2*Witer
    Riter = h*Svel(Sub+Riter, 1, SubE+Witer)
    Witer = h*SEvel(Sub+Riter,1, SubE+Witer)
    CounterW+=Witer
    CounterR+=Riter
    SubE= SubE + CounterW/6
    SEaxis.append(SubE)
    Sub= Sub + CounterR/6 -k3*h*Sub
    Saxis.append(Sub)
    Eaxis.append(1-SubE)
    Pr=((1-Sub)/2-SubE)
    Prnew=((1-Sub)/2-SubE)-k3*h*Pr
    Paxis.append(Prnew)
    Pr=0
    Prnew=0
    CounterW = 0
    CounterR = 0
# Plot two curves
plt.plot(taxis, Saxis, label="S", color='blue')  # First curve in blue
plt.plot(taxis, SEaxis, label="SE", color='red')  # Second curve in red
plt.plot(taxis, Eaxis, label="E", color='yellow')
plt.plot(taxis, Paxis, label="P", color='green')
# Labels and title
plt.xlabel("Time")
plt.ylabel("Concentration")
plt.title("2S Michaelis-Menten")
plt.legend()  # Show legend
```

# **Тепер реалізую метод Адамса Башфорта 4 порядку**
Розганяємо його послідовно методами адамса башфорта від 1 до 3 порядку, цікаво зазначити що метод адамса башфорта першого порядку є просто методом ейлеру


```python
#Блок пам'яті точок функції
Subi = 1
Subii = 0
Subiii = 0
Subiiii = 0
SubEi = 0
SubEii = 0
SubEiii = 0
SubEiiii = 0

Eaxis = []
Paxis = []
Saxis=[]
SEaxis=[]

CounterS = 0
CounterSE = 0
h = 0.0001

taxis=list(i*h for i in range(500008))
SEaxis.append(SubEi)
Saxis.append(Subi)
Eaxis.append(1-SubEi)
Paxis.append((1-Subi)/2-SubEi)

Subii = Subi + h*Svel(Subi, 1, SubEi)
SubEii = SubEi + h*SEvel(Subi, 1, SubEi)
SEaxis.append(SubEii)
Saxis.append(Subii)
Eaxis.append(1-SubEii)
Paxis.append((1-Subii)/2-SubEii)

Subiii = Subii + h*(3/2*Svel(Subii, 1, SubEii)-1/2*Svel(Subi, 1, SubEi))
SubEiii = SubEii + h*(3/2*SEvel(Subii, 1, SubEii)-1/2*SEvel(Subi, 1, SubEi))
SEaxis.append(SubEiii)
Saxis.append(Subiii)
Eaxis.append(1-SubEiii)
Paxis.append((1-Subiii)/2-SubEiii)

Subiiii = Subiii + h*(23/12*Svel(Subiii, 1, SubEiii)-4/3*Svel(Subii, 1, SubEii)+5/12*Svel(Subi, 1, SubEi))
SubEiiii = SubEiii + h*(23/12*SEvel(Subiii, 1, SubEiii)-4/3*SEvel(Subii, 1, SubEii)+5/12*SEvel(Subi, 1, SubEi))
SEaxis.append(SubEiiii)
Saxis.append(Subiiii)
Eaxis.append(1-SubEiiii)
Paxis.append((1-Subiiii)/2-SubEiiii)


for i in range(500004):
    CounterS = h*(55/24*Svel(Subiiii, 1, SubEiiii)-59/24*Svel(Subiii, 1, SubEiii)+37/24*Svel(Subii, 1, SubEii)-3/8*Svel(Subi, 1, SubEi))
    CounterSE = h*(55/24*SEvel(Subiiii, 1, SubEiiii)-59/24*SEvel(Subiii, 1, SubEiii)+37/24*SEvel(Subii, 1, SubEii)-3/8*SEvel(Subi, 1, SubEi))
    
    Subi = Subii
    Subii = Subiii
    Subiii = Subiiii
    Subiiii = Subiiii + CounterS
    
    SubEi = SubEii
    SubEii = SubEiii
    SubEiii = SubEiiii
    SubEiiii = SubEiiii + CounterSE
    SEaxis.append(SubEiiii)
    Saxis.append(Subiiii)
    Eaxis.append(1-SubEiiii)
    Paxis.append((1-Subiiii)/2-SubEiiii)
    
# Plot two curves
plt.plot(taxis, Saxis, label="S", color='blue')  # First curve in blue
plt.plot(taxis, SEaxis, label="SE", color='red')  # Second curve in red
plt.plot(taxis, Eaxis, label="E", color='yellow')
plt.plot(taxis, Paxis, label="P", color='green')
# Labels and title
plt.xlabel("Time")
plt.ylabel("Concentration")
plt.title("2S Michaelis-Menten")
plt.legend()  # Show legend
```
