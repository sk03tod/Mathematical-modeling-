```python
import numpy as np
import matplotlib.pyplot as plt
import random as rnd
```

```python
def Energy_izing(x:list,topology="limit",J=-4)->float:
    counter=0
    if topology=="limitless":
        for i in range(len(x)):
            counter+=x[i%len(x)]*x[(i+1)%len(x)]
    elif topology=="limit":
        for i in range(len(x)-1):
            counter+=x[i]*x[i+1]
    return -1*J*counter
```

```python
kB=8.6e-5
Temp=30
N=100
indexlist=list(i for i in range(N))
```

```python
Energy_reccord=[]
Spins=list(1 for i in range(N))
for i in range(16*N):
    a = rnd.choice(indexlist)
    E0=Energy_izing(Spins)
    Spins[a]=-1*Spins[a]
    E1=Energy_izing(Spins)
    if E1>=E0 and rnd.random()>np.exp((E0-E1)/(kB*Temp)):
        Spins[a]=-1*Spins[a]
    Energy_reccord.append(Energy_izing(Spins))
```

```python
Energy_reccord_cycl=[]
Spins=list(1 for i in range(N))
for i in range(16*N):
    a = rnd.choice(indexlist)
    E0=Energy_izing(Spins, topology="limitless")
    Spins[a]=-1*Spins[a]
    E1=Energy_izing(Spins, topology="limitless")
    if E1>=E0 and rnd.random()>np.exp((E0-E1)/(kB*Temp)):
        Spins[a]=-1*Spins[a]
    Energy_reccord_cycl.append(Energy_izing(Spins, topology="limitless"))
```

```python
plt.plot(Energy_reccord, color="red")
plt.plot(Energy_reccord_cycl)
plt.show()
```
