```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Функції перших похідних
def rabbit_der(r, w, a, b):
    rabderiv = a*r-b*w*r
    return rabderiv
def wolf_der(r, w, g, s):
    wolderiv = s*r*w-g*w
    return wolderiv
```

```python
Waxis=[]
Raxis=[]
taxis=list(i/20 for i in range(1, 2001))
Wol = 2
Rab = 10
Witer = 0
Riter = 0
CounterW = 0
CounterR = 0
h = 0.1
for i in range(2000):
    Riter = h*rabbit_der(Rab, Wol, 0.3, 0.06)
    Witer = h*wolf_der(Rab, Wol, 0.4, 0.05)
    CounterW+=Witer
    CounterR+=Riter
    Riter = h*rabbit_der(Rab+Riter/2, Wol+Witer/2, 0.3, 0.06)
    Witer = h*wolf_der(Rab+Riter/2, Wol+Witer/2, 0.4, 0.05)
    CounterR+=2*Riter
    CounterW+=2*Witer
    Riter = h*rabbit_der(Rab+Riter/2, Wol+Witer/2, 0.3, 0.06)
    Witer = h*wolf_der(Rab+Riter/2, Wol+Witer/2, 0.4, 0.05)
    CounterR+=2*Riter
    CounterW+=2*Witer
    Riter = h*rabbit_der(Rab+Riter, Wol+Witer, 0.3, 0.06)
    Witer = h*wolf_der(Rab+Riter, Wol+Witer, 0.4, 0.05)
    CounterW+=Witer
    CounterR+=Riter
    Wol= Wol + CounterW/6
    Waxis.append(Wol)
    Rab= Rab + CounterR/6
    Raxis.append(Rab)
    CounterW = 0
    CounterR = 0
plt.plot(taxis, Waxis)
plt.xlabel('time')
plt.ylabel('Wolves')
plt.title('Wolves')
plt.show()
plt.plot(taxis, Raxis)
plt.xlabel('time')
plt.ylabel('Rabbits')
plt.title('Rabbits')
plt.show()
plt.plot(Waxis, Raxis)
plt.xlabel('Wolves')
plt.ylabel('Rabbits')
plt.title('Phase space')
plt.show()
```
