```python
import numpy as np 
import matplotlib.pyplot as plt
import random as rnd
```

```python
Walkingarea = 20
l = 0.2
```

```python
Xgrid = list(round(0.2*i, 2) for i in range(int(Walkingarea/l)))
print(Xgrid)
```

```python
def Random_walking_2D(ntr, nstp):
    array = list(list(0 for i in range(int(Walkingarea/l))) for j in range(int(Walkingarea/l)))
    for i in range(ntr):
        counter1 = int(Walkingarea/2/l)
        counter2 = int(Walkingarea/2/l)
        for j in range(nstp):
            a = rnd.random()
            if a<0.25: counter1+=1
            elif 0.25<a<0.5: counter1-=1
            elif 0.5<a<0.75: counter2+=1
            elif 0.75<a: counter2-=1
        if counter1<len(array[0]) and counter2<len(array[0]):
            array[counter1][counter2]+=1
    return np.array(array)/ntr
```

```python
Xc, Yc= np.meshgrid(Xgrid, Xgrid)
plt.pcolormesh(Xc, Yc, Random_walking_2D(10000, 10), cmap='hot')
plt.show()

plt.pcolormesh(Xc, Yc, Random_walking_2D(10000, 30), cmap='hot')
plt.show()

plt.pcolormesh(Xc, Yc, Random_walking_2D(10000, 100), cmap='hot')
plt.show()

plt.pcolormesh(Xc, Yc, Random_walking_2D(10000, 200), cmap='hot')
plt.show()

plt.pcolormesh(Xc, Yc, Random_walking_2D(1000000, 700), cmap='hot')
plt.show()
```
